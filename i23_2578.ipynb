## BiLSTM vs Transformer Comparison

The strongest BiLSTM result in this repo is the fine-tuned POS sequence labeler, which achieved token accuracy of 0.9864 and macro-F1 of 0.9740, while the Transformer topic classifier achieved test accuracy of 0.5111 and macro-F1 of 0.3505 on the 5-way article classification task. If we compare the headline accuracy numbers directly, the BiLSTM is higher by 0.4753, although the two models were trained on different supervised objectives. In terms of convergence, the BiLSTM reached very strong validation performance much earlier on its task, while the Transformer needed most of the 20 scheduled epochs before its validation curve stabilized. The Transformer also showed a long flat phase at the beginning, which suggests it needed more optimization steps before it started separating classes reliably.

Per epoch, the Transformer is generally more expensive when sequence length is fixed at 256 because self-attention compares every token with every other token, so its cost grows quadratically with sequence length. The BiLSTM processes tokens recurrently, which is sequential, but on this small dataset and shorter labeled sequences it remained a more data-efficient choice. The attention heatmaps from the final encoder layer show that the Transformer does not attend uniformly: it places stronger weight on the CLS token, topical anchor words, and prominent country or sports terms. In the correctly classified sports articles, the model gives noticeably stronger attention to match or team-related tokens, while in international articles it concentrates more on country names and geopolitical markers.

Those heatmaps suggest that the Transformer is learning a topic-summarization behavior rather than tracking every token equally. That is useful for document classification, but it also means performance depends heavily on whether a small training set exposes enough repeated topic cues. With only about 200 to 300 articles, the more appropriate architecture is usually the BiLSTM when the task is narrow and labels are limited, because it tends to need less data to generalize. The Transformer becomes more attractive when more labeled documents are available, especially if we want richer long-range interactions and interpretable attention maps. In this project, the BiLSTM was the safer low-data model, while the Transformer gave more interpretable document-level focus patterns but lower end accuracy on the available split.
